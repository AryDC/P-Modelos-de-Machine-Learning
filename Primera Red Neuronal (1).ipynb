```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Celsius = np.linspace(-273, 273.15, num=10000)
Farenheit = np.array([((temperatura*1.8)+32) for temperatura in Celsius])
visualizacion = pd.DataFrame(columns=["Celsius", "Farenheit"])
visualizacion["Celsius"] = pd.Series(Celsius)
visualizacion["Farenheit"] = pd.Series(Farenheit)
visualizacion.plot()
```

```python
capa_entrada = tf.keras.layers.Dense(units=3, input_shape=[1])
capa2 = tf.keras.layers.Dense(units=3)
capa_salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([capa_entrada, capa2, capa_salida])
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = "mean_squared_error"
)
```

```python
historial = modelo.fit(Celsius, Farenheit, epochs=100, verbose=False)
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Longitud de Perdida")
plt.plot(historial.history["loss"])
```

```python
resultado = modelo.predict([100])
print(f"El resultado es {resultado} Farenheit")
```
